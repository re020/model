```python
import os
import pandas as pd
import numpy as np
from rdkit.Chem import AllChem,DataStructs
from rdkit.Chem import MACCSkeys
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
```

```python
def get_features(mol):
    # 初始化存储指纹值的字典
        data = {}

        # MACCS键型指纹
        fp_maccs = MACCSkeys.GenMACCSKeys(mol)
        arr_maccs = np.zeros((1, ), dtype=np.int8)
        DataStructs.ConvertToNumpyArray(fp_maccs, arr_maccs)
        for i in range(len(arr_maccs)):
            data[f'Analyte_maccs_fingerprint_{i}'] = arr_maccs[i]

        # 摩根指纹
        fp_Morgan = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)
        arr_Morgan = np.zeros((1, ), dtype=np.int8)
        DataStructs.ConvertToNumpyArray(fp_Morgan, arr_Morgan)
        for i in range(len(arr_Morgan)):
            data[f'Analyte_Morgan_fingerprint_{i}'] = arr_Morgan[i]

        # 计算拓扑扭曲指纹
        fp_top = AllChem.GetHashedTopologicalTorsionFingerprintAsBitVect(mol)
        arr_top = np.zeros((1, ), dtype=int)
        DataStructs.ConvertToNumpyArray(fp_top, arr_top)
        for i in range(len(arr_top)):
            data[f'Analyte_topological_torsion_{i}'] = arr_top[i]

        return data
```

```python
folder_path = r"C:\Users\re\Desktop\data\dataset\fenli\huan_RS"
# 初始化空的DataFrame
columns = ['Analyte_maccs_fingerprint','Analyte_topological_torsion','Analyte_Morgan Fingerprint']
df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith(".sdf"):
        file_path = os.path.join(folder_path, filename)
        mol = Chem.MolFromMolFile(file_path, sanitize=False)  # 读取分子，不进行sanitize

        if mol is None:
            print(f"Failed to load molecule from {filename}")
            continue

        # 手动初始化环信息
        mol.UpdatePropertyCache(strict=False)
        Chem.GetSymmSSSR(mol)  # 计算并储存环信息

        # 计算特征
        data_r = get_features(mol)
        # 将结果添加到DataFrame中
        df = df._append(pd.Series(data_r, name=filename[:-4]))

# 如果需要，可以保存DataFrame到CSV文件
print(df.columns)
df=df.drop(['Analyte_maccs_fingerprint','Analyte_topological_torsion','Analyte_Morgan Fingerprint'],axis=1)
df.to_csv('fenli_huan_print.csv')
```
